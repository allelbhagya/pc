### taking a test sample

```python
X,y = iris.data, iris.target

test = np.array([5.4, 2.2, 4.1, 1.2])
test
```

### dist dictionary to save the distances of every data point from test sample

```python
dist = []

for i in range(len(X)):
    dist.append(np.sqrt( ((X[i][0]-test[0])**2) + ((X[i][1]-test[1])**2) + ((X[i][2]-test[2])**2) + ((X[i][3]-test[3])**2)))
len(dist)
```

```python
idx = {}
for i in range(len(dist)):
    idx[dist[i]] = i
```

### taking k = 5

```python
md = np.sort(dist)

k = 5
k_0 = k_1 = k_2 = 0

for i in range(k):
    if y[idx[md[i]]] ==0:
        k_0+=1
    elif y[idx[md[i]]] == 1:
        k_1+=1
    else:
        k_2+=1
k_0, k_1, k_2
```
